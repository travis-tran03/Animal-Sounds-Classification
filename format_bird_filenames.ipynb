## Goal: Rename bird song data by label

```python
import os
import pandas as pd

# Paths
csv_file_path = 'birdsong/birdsong_metadata.csv'  # Replace with the path to your .csv file
audio_folder_path = 'birdsong/songs/songs'  # Replace with the path to your audio files folder

# Read the CSV file
df = pd.read_csv(csv_file_path)

# Create a dictionary mapping IDs to bird breeds
id_to_breed = dict(zip(df['file_id'], df['english_cname']))

# Iterate through the audio files in the folder
for filename in os.listdir(audio_folder_path):
    if filename.endswith('.flac'):  # Adjust extensions as needed
        # Extract the ID from the filename (skipping 'xc')
        file_id = int(os.path.splitext(filename)[0][2:])

        # Get the corresponding breed from the dictionary
        if file_id in id_to_breed:
            breed = id_to_breed[file_id]
            
            # Format the breed name: lowercase and replace spaces with underscores
            #formatted_breed = f"bird_{breed.lower().replace(' ', '_')}"
            #print(formatted_breed)
            
            # Create the new filename
            new_filename = f"{breed}_{file_id}{os.path.splitext(filename)[1]}"
            # Rename the file
            os.rename(
                os.path.join(audio_folder_path, filename),
                os.path.join(audio_folder_path, new_filename)
            )
            print(f"Renamed {filename} to {new_filename}")
        else:
            print(f"No breed found for ID {file_id} in the CSV file.")
            print(type(file_id))
```

```python
import os

# Path to your audio files folder
audio_folder_path = 'birdsong/songs/songs'  # Replace with the path to your audio files folder

# Iterate through the audio files in the folder
for filename in os.listdir(audio_folder_path):
    if filename.endswith('.wav') or filename.endswith('.mp3') or filename.endswith('.flac'):  # Include .flac
        # Split the filename into breed and ID
        parts = os.path.splitext(filename)[0].split('_')  # Split on underscore
        if len(parts) >= 2:  # Ensure the filename has the expected format (breed_ID)
            breed = '_'.join(parts[:-1])  # Join all parts except the last one (ID)
            file_id = parts[-1]  # The last part is the ID

            # Format the breed name: lowercase and replace spaces with underscores, and add 'bird_' prefix
            formatted_breed = f"bird_{breed.lower().replace(' ', '_')}"
            # Create the new filename
            new_filename = f"{formatted_breed}_{file_id}{os.path.splitext(filename)[1]}"
            # Rename the file
            os.rename(
                os.path.join(audio_folder_path, filename),
                os.path.join(audio_folder_path, new_filename)
            )
            print(f"Renamed {filename} to {new_filename}")
        else:
            print(f"Filename {filename} does not match the expected format (breed_ID).")
```
